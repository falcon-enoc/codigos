```python
import numpy as np
import sklearn as sk
```

```python
num_points=[int(1e3),int(1e4),int(1e6)]
num_dims=[32,64,128]
num_leaf=[40,80,120]
```

```python
import faiss
```

```python
import time 
from sklearn.neighbors import KDTree

query_data=list()
build_data=list()
leaf_data=list()
dims_data=list()
points_data=list()
for n in num_points:
    for d in num_dims:
        for j in num_leaf:
            X=np.random.random((n,d))
            query=np.random.random(d)
            t1=time.time()
            tree = KDTree(X, leaf_size=j)
            t2=time.time()
            build_data.append(t2-t1)
            t3=time.time()
            t4=time.time()
            ind = tree.query(query.reshape(1,-1), k=1,return_distance=False,dualtree=True)
            t5=time.time()
            query_data.append(t5-t4)
            dims_data.append(d)
            points_data.append(n)
            leaf_data.append(j)
            
            
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'build':build_data,'query':query_data,'dims':dims_data,
                 'points':points_data,'leaf_size':leaf_data})
```

```python
df.groupby(['dims','points','leaf_size']).mean()
```

```python
import time

query_data=list()
build_data=list()
leaf_data=list()
dims_data=list()
points_data=list()
for n in num_points:
    for d in num_dims:
        X=np.random.random((n,d))
        query=np.random.random(d)
        t1=time.time()
        index = faiss.IndexFlatL2(d)
        index.add(X) 
        t2=time.time()
        build_data.append(t2-t1)
        t3=time.time()
        t4=time.time()
        D, I = index.search(query.reshape(1,-1), k=1)             
        t5=time.time()
        query_data.append(t5-t4)
        dims_data.append(d)
        points_data.append(n)
```

```python
import pandas as pd

df_faiss=pd.DataFrame({'build':build_data,'query':query_data,'dims':dims_data,
                 'points':points_data})
```

```python
df_faiss.groupby(['dims','points']).mean()
```
